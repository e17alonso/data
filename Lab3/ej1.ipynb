```python
pregunta1='''
SELECT 
    a.first_name,
    a.last_name,
    b.emp_no,
    b.title,
    c.from_date,
    c.to_date,
    c.salary
FROM 
    employee.titles AS b
LEFT JOIN 
    employee.salaries AS c ON b.emp_no = c.emp_no
LEFT JOIN 
    employee.employees AS a ON b.emp_no = a.emp_no
WHERE 
    c.emp_no = 12557 
ORDER BY 
    c.from_date;
'''
pd.read_sql(pregunta1,connection)
```

```python
pregunta2='''
SELECT
    a.emp_no,
    a.birth_date,
    a.first_name,
    a.last_name,
    b.dept_no,
    b.from_date,
    b.to_date,
    c.dept_name
FROM
    employee.employees a
JOIN
    employee.dept_emp b ON a.emp_no = b.emp_no
JOIN
    employee.departments c ON b.dept_no = c.dept_no
WHERE
    a.emp_no = 12557
LIMIT 1;
'''
pd.read_sql(pregunta2,connection)
```

```python
pregunta3='''
SELECT DISTINCT 
    e.emp_no,
    e.birth_date,
    e.first_name,
    e.last_name,
    e.gender,
    e.hire_date,
    MAX(s.salary) OVER(PARTITION BY e.emp_no) AS max_salary
FROM 
    employee.employees e
LEFT JOIN 
    employee.salaries s ON e.emp_no = s.emp_no
ORDER BY 
    max_salary;
'''
pd.read_sql(pregunta3,connection)
```
